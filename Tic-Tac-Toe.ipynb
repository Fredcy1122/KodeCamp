```python
def play(board):
    # Define winning combinations
    wins = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # Rows
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # Columns
        [0, 4, 8], [2, 4, 6]              # Diagonals
    ]

    # Determine current player
    x_count = board.count('x')
    o_count = board.count('o')
    player = 'x' if x_count == o_count else 'o'
    opponent = 'o' if player == 'x' else 'x'

    # Helper function to find a winning move
    def find_winning_move(symbol):
        for combo in wins:
            values = [board[i] for i in combo]
            if values.count(symbol) == 2 and values.count('') == 1:
                return combo[values.index('')]
        return None

    # 1️⃣ Try to win
    move = find_winning_move(player)
    if move is not None:
        return move

    # 2️⃣ Block opponent's winning move
    move = find_winning_move(opponent)
    if move is not None:
        return move

    # 3️⃣ Take center if available
    if board[4] == '':
        return 4

    # 4️⃣ Take a corner if available
    for i in [0, 2, 6, 8]:
        if board[i] == '':
            return i

    # 5️⃣ Take any side
    for i in [1, 3, 5, 7]:
        if board[i] == '':
            return i

    # Fallback — choose any empty square
    for i in range(9):
        if board[i] == '':
            return i
```

```python
board = ['', '', '', '', '', '', '', '', '']
print(play(board))  
```
