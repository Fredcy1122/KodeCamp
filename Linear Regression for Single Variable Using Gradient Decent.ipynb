```python
# prediction

def pred (x, m, c):
    y = m*x + c
    
    return y
```

```python
# funtion to compute mean square error

def mse (y_true, y_pred):
    # MSE is the average of the diff between actual value and pred value this is squared
    se = np.mean((y_true - y_pred) ** 2)
    
    return se
```

Gradient Descent is an optimization algorithm used to train machine learning models by minimizing the loss function.




```python
# to get the gradient, we find the gradient with respect to slope(dm) and intercept(dc)
# dj/dm = (-2/n) * sum(x_i * (error)) note: error = y_true - y_pred
# dj/dc = (-2/n) * sum(error)

def gradient (x, y, m, c):
    
    n = x.size  #size of the data
    y_hat = m*x + c
    error = y - y_hat
    dm = (-2.0/n) * np.sum(x * error)
    dc = (-2.0/n) * np.sum(error)
    
    return dm, dc
```

```python
lr = 0.01  #learning rate
m = 0
c = 0

errors = []
ms = []
cs = []
```

```python
for a in range (10): # a loop to have multiple itirations
    y_hat = pred(x, m, c)  #cal predicted values
    error = mse(y, y_hat)  #cal the cost function
    grad = gradient(x, y, m, c) #cal the gradrient
    
    errors.append(error)
    ms.append(m)
    cs.append(c)
    
    #adjust parameters
    m = m - lr * grad[0]
    c = c - lr * grad[1]
    
    print(f"The MSE is: {error}")
    
    
    
```
